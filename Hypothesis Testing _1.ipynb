## 4.Plotting Data

```python
plt.subplots(figsize = (16,6))
plt.subplot(141)
plt.boxplot(lab['Laboratory 1'])
plt.title('Laboratory 1')
plt.subplot(142)
plt.boxplot(lab['Laboratory 2'])
plt.title('Laboratory 2')
plt.subplot(143)
plt.boxplot(lab['Laboratory 3'])
plt.title('Laboratory 3')
plt.subplot(144)
plt.boxplot(lab['Laboratory 4'])
plt.title('Laboratory 4')
plt.show()
```

```python
plt.figure(figsize = (8,6))
labels = ['Lab 1', 'Lab 2','Lab 3', 'Lab 4']
sns.distplot(lab['Laboratory 1'])
sns.distplot(lab['Laboratory 2'])
sns.distplot(lab['Laboratory 3'])
sns.distplot(lab['Laboratory 4'])
plt.legend(labels)
```

## 5.ANOVA Test

```python
statistic,pvalue=stats.f_oneway(lab['Laboratory 1'],lab['Laboratory 2'],lab['Laboratory 3'],lab['Laboratory 4'])
print('statistics is {}\npvalue is {}'.format(statistic,pvalue))
```

```python
alpha=0.05
if (pvalue<=alpha):
    print('There is significant difference in average TAT of reports between the laboratories (Ha)')
else:
    print('There is no significant change in verage TAT of reports between the laboratories (Ho)')
```

## There is significant difference in average TAT of reports between the laboratories (Ha) - We are Rejecting Null Hypothesis

## We are going to conduct a Test of Independence using Chi-Square  test with Contingency table

## 1.Importing data

```python
Sales_data=pd.read_csv('BuyerRatio.csv')
```

```python
Sales_data
```

## 2.Chi-Square test

```python
from scipy.stats import chi2_contingency
```

```python
Sales_data=[[50,435],[142,1523],[131,1356],[70,750]]
```

```python
stat, p_value, dof, exp = stats.chi2_contingency(Sales_data)
print('statistics is {}\npvalue is {}\ndof is {}\nexpected value is{}'.format(stat,p_value,dof,exp))
```

```python
alpha=0.05
if p_value <= alpha:
    print('We reject Null Hypothesis there is a significance difference between TAT of reports of the laboratories')
else:
    print('We fail to reject Null hypothesis')
```

### We fail to reject Null Hypothesis. Therefore, there is no association or dependency between male-female buyers rations and are similar across regions. Hence, Independent samples
